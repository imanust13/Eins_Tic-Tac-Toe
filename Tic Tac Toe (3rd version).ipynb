```python
## Life is too short to debug


# FUNCTIONS PART


# eins: prints out a board
from IPython.display import clear_output # this only works in Jupyter notebook
def board_display(board):
    
    clear_output() # this only works in Jupyter notebook
    print(board[7]+'|'+board[8]+'|'+board[9])
    print(board[4]+'|'+board[5]+'|'+board[6])
    print(board[1]+'|'+board[2]+'|'+board[3])
    
# zwei: takes in a player input and assigns their marker as 'X' or 'O'
def player_marker():
    marker=''
    while marker!='X' and marker!='O':
        marker=input('idol, choose a marker, X or O: ')
        if marker=='X':
            return ('X','O')
        else: 
            return ('O','X')
        
# drei: takes in the board list object, a marker, and a desired position, and assigns it to the board
def place_marker(board,marker,position):
    board[position]=marker
    return board

# vier: takes in a board and a marker, and then checks to see if that marker has won
def win_check(board,marker):
        return board[1]==board[2]==board[3]==marker or\
           board[4]==board[5]==board[6]==marker or\
           board[7]==board[8]==board[9]==marker or\
           board[1]==board[4]==board[7]==marker or\
           board[2]==board[5]==board[8]==marker or\
           board[3]==board[6]==board[9]==marker or\
           board[1]==board[5]==board[9]==marker or\
           board[3]==board[5]==board[7]==marker 
    # it works, but certainly not elegant, as listing all winning scenarios is far from a scalable approach.
    
# f√ºnf: decides which player goes first
import random 
def play_first():
    if random.randint(1,2)==1:
        return 'idol'
    else:
        return 'tsu'

# sechs: checks if a position on the board is available
def free_position(board,position):
    return board[position]==' '

# sieben: checks if the board is full
def full_board_check(board):
    return ' ' not in board

# acht: asks for a player's next move
def make_a_move(board):
    position=' '
    while position not in list(range(1,10)) or not free_position(board,position):
        position=int(input('What is your next move, 1-9: '))
    return position

# neun: asks the player if they want to replay
def replay():
    if input('Play again? Y or N: ')=='Y':
        return True
    else:
        return False
```

```python
# GAME PLAY PART

# my board must be ['#',' ',' ',' ',' ',' ',' ',' ',' ',' '] for some functions to work
# I need a BIG while loop to keep running the game, and a NESTED while loop to shift between the two players

print('Welcome to Tic Tac Toe!')

# I had to put this who plays first part outside of the while loop
# otherwise the clear_output function clears everything before who goes first can be seen

turn=play_first()
print(turn+' goes first')

while True:
    board=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    
# set up the game:

# ask the player to choose a marker & tuple unpacking
    idol_marker,tsu_marker=player_marker()

    game_on=True
    
# actual game play 
    while game_on:
        if turn=='idol': # if it's idol's turn, this block of code will be executed. 
            # show the board
            board_display(board)
            
            # ask idol to choose a position (this function already checks if this position if available)
            # it returns a position (1-9) for later use
            position=make_a_move(board)
            
            # update the board
            place_marker(board,idol_marker,position)
            
            # check if the updated board is won for idol
            if win_check(board,idol_marker):
                board_display(board)
                print('idol has won!')
                
                game_on=False
            
            # check if the board is full
            else: 
                if full_board_check(board):
                    board_display(board)
                    print("It's a draw!")
                    
                    game_on=False # or break
                else:
                    turn='tsu'
            
            # if not win and there're still other free positions, tsu's turn
        else: # which means if turn=='tsu'
            # show the board
            board_display(board)
            
            # ask tsu to choose a position (this function already checks if this position if available)
            # it returns a position (1-9) for later use
            position=make_a_move(board)
            
            # update the board
            place_marker(board,tsu_marker,position)
            
            # check if the updated board is won for idol
            if win_check(board,tsu_marker):
                board_display(board)
                print('tsu has won!')
                
                game_on=False
            
            # check if the board is full
            else: 
                if full_board_check(board):
                    board_display(board)
                    print("It's a draw!")
                    
                    game_on=False # or break
                else:
                    turn='idol'
            

    if not replay():
        break # break out of the BIG while loop and the game terminates
```
